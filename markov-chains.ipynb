```python
def plot(i, arr):
    fig, axs = plt.subplots(1, 1, sharey=True, tight_layout=True)
    plt.axis([-6, 6, 0, 1])
    axs.bar(range(-5, 6, 1), height=arr)
    plt.savefig(f'{i}.png')
```

```python
for i in range(0, 30, 1):
    plot(i, np.matmul(x0, np.linalg.matrix_power(T, i)))
```
